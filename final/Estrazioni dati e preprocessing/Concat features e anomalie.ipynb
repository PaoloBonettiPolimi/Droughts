prendo i dati di target e le features e faccio un merge per avere il dataset finale composto da anomalie NDVI + anomalie features

```python
import pandas as pd

features = pd.read_pickle('features.pickle')
```

```python
df = pd.read_csv('cultivable_ndvi.csv')
```

```python
features = features.iloc[1:].reset_index(drop=True)
```

```python
df = df.iloc[:1038]
```

```python
features
```

```python
df = pd.concat([df,features.iloc[:, 2:]], axis=1)
```

```python
df.drop(df.filter(regex='NDVI').columns.tolist(), axis=1, inplace=True)
```

```python
df = df[['Date','Year','Week'] + df.filter(regex='anomalia|Anomalia').columns.tolist()]
```

```python
df.to_csv('cultivable_solo_anomalie.csv', index=False)
```
