```python
import src.preprocessing_functions as pp
import src.mixedRVMI as mx
import pandas as pd 
import pickle
from scipy.stats import pearsonr
import numpy as np
```

```python
ndvi_file="ndvi+a_1981-2021.pickle"
pt_file="piogge_temp_aggregati.pickle"

ndvi = pd.read_pickle(ndvi_file)
pt = pd.read_pickle(pt_file)
```

```python
df = pd.merge(pt.iloc[:, 1:],ndvi, on=['Year','Week'])
```

Pickle con target *anomalia nord*

```python
features = df.iloc[:, 0:20].values
target = df.iloc[:, 23].values
```

```python
output = { "X" : features, "Y" : target}

with open("dataset_target_nord.pickle", "wb") as f:
    pickle.dump(output, f)
```

Pickle con target *anomalia sud*

```python
target = df.iloc[:, 24].values

output = { "X" : features, "Y" : target}

with open("dataset_target_sud.pickle", "wb") as f:
    pickle.dump(output, f)
```

Pickle con target *anomalia all*

```python
target = df.iloc[:, 25].values

output = { "X" : features, "Y" : target}

with open("dataset_target_all.pickle", "wb") as f:
    pickle.dump(output, f)
```

# Calcolo MI tra features e target

```python
pp_features = pp.data_scale(features,flag="ScaleAndNoise") # modifica anche il valore di features, che vado a ripristinare
features = df.iloc[:, 0:20].values
```

```python
target_nord = df.iloc[:, 23].values
scores = mx.estimateAllMI(pp_features,pp.data_scale(target_nord, flag="ScaleAndNoise"), k=len(features)//20)

list(zip([num[1] for num in scores],df.columns.tolist()[0:20]))
```

```python
target_sud = df.iloc[:, 24].values
scores = mx.estimateAllMI(pp_features,pp.data_scale(target_sud, flag="ScaleAndNoise"), k=len(features)//20)

list(zip([num[1] for num in scores],df.columns.tolist()[0:20]))
```

```python
target_all = df.iloc[:, 24].values
scores = mx.estimateAllMI(pp_features,pp.data_scale(target_all, flag="ScaleAndNoise"), k=len(features)//20)
list(zip([num[1] for num in scores],df.columns.tolist()[0:20]))
```

# Calcolo correlazione lineare tra ciascuna feature con il target

```python
rho_scores = []
target_nord = df.iloc[:,23].values # lo ripristino dopo il preprocessing

for col in range(features.shape[1]):
    rho_scores.append(pearsonr(features[:, col], target_nord)[0])

list(zip(rho_scores, df.columns.tolist()[0:20]))
```

```python
rho_scores = []

target_sud = df.iloc[:,24].values

for col in range(features.shape[1]):
    rho_scores.append(pearsonr(features[:, col], target_sud)[0])

list(zip(rho_scores, df.columns.tolist()[0:20]))
```

```python
rho_scores = []

target_all = df.iloc[:,25].values

for col in range(features.shape[1]):
    rho_scores.append(pearsonr(features[:, col], target_all)[0])

list(zip(rho_scores, df.columns.tolist()[0:20]))
```
